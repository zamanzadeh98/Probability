## Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

import warnings
warnings.simplefilter("ignore")
```

## train test split

```python
x_train, x_test, y_train, y_test = train_test_split(data.data, 
                                                    data.target,
                                                    test_size=0.2)
print("# Train data points: ", x_train.shape[0])
print("# test data points: ", x_test.shape[0])
```

## model

```python
class LogisticRegression:
    
    def __init__(self, lr=0.010, num_iter=300):
        """
        Logistic Regression model
        """
        self.lr = lr
        self.num_iter = num_iter 
    

    
    def sigmoid(self, z):
        return 1/ (1+ np.exp(-z))
                   
                   
    def fit(self, x, y):
        # we need to initialize our parameters
        #(weights and bias) to store the new ones
        # in them
        
        numSamples = x.shape[0]
        self.weights = np.random.random(x.shape[1])
        # self.weights = np.zeros(x.shape[1])
        self.bias = np.random.random(1)
        
                   
        # Gradient descent
        for iter in range(self.num_iter):
            
           # linear model part (calculating z)
            linearTransfrom = np.dot(x, self.weights) + self.bias
            z = self.sigmoid(linearTransfrom)

           # gradient calculation
            dw = (1 / numSamples) * np.dot(x.T, (z - y))
            db = (1 / numSamples) * np.sum(z - y)

            # update parameters
            self.weights -= self.lr * dw
            self.bias -=    self.lr * db
        
        
    def predict(self, x_test):
        linearTransfrom = np.dot(x_test, self.weights) + self.bias
        z = self.sigmoid(linearTransfrom)
        # Thresholding by comparing with 0.5
        yPred = [1 if i > 0.5 else 0 for i in z]

        return yPred

        

def accuracyCal(yPred, yTrue):
    correct_predictions = np.sum(yPred == yTrue)
    total_samples = len(yTrue)
    accuracy = correct_predictions / total_samples * 100
    return accuracy
```

```python
accs = []

for j in range(100):
    model = LogisticRegression()
    model.fit(x_train, y_train)

    yPred = model.predict(x_test)
    accs.append(accuracyCal(yPred, y_test))
```

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.set_title("Accuracy distribution with different initialization")
sns.kdeplot(accs, color="purple")
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.set_xlabel("accuracy")

plt.show()
```
