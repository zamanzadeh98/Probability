# Beta

**Question**
Consider a machine learning model designed for disease diagnosis. It has successfully classified 60 individuals accurately and made incorrect classifications for 20 individuals. What is the probability of it making a correct classification?

**Answer**

C: correctly calssified
F: Failed
we can easily opt frequentists appoach and use the folmula:

$ P(C) = \frac{\text{# C}}{\text{# F}} $

However, relying on a single number to express our belief might not be advisable. Instead, we can utilize the beta distribution to capture our uncertainty regarding the true probability of success. By employing the beta distribution, we can treat \(P(C)\) as a continuous random variable.

So, what we are looking for is : $ P(X=x| C=60, F=20) = ? $

here we assume the model was unbiased therefore, the prior can be a uniform distribution:

$ f(x| \alpha, \beta) = \frac{x^{\alpha-1} (1-x)^{\beta-1}}{B(\alpha, \beta)} $

where
$ B(a, b) = \int_0^1 t^{a-1} (1-t)^{b-1} \, dt  $


```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
a = 61
b = 21

dist = beta(a, b)

arr = np.linspace(0, 1, 100)
probs = dist.pdf(arr)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))

ax[0].plot(arr, [1 for x in arr], color='black')
ax[0].spines["right"].set_visible(False)
ax[0].spines["top"].set_visible(False)
ax[0].set_title("prior distribution")

ax[1].plot(arr, probs, color='black')
ax[1].spines["right"].set_visible(False)
ax[1].spines["top"].set_visible(False)
ax[1].set_title("Posterior distribution")


ax[0].set_ylabel("PDF")

plt.show()
```
