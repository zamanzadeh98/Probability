# Normal random variable

Let X be a random variable which is represented by a Mean and a variance:

$$X \sim \mathcal{N}(\mu, \sigma^2)$$

Normal PDF is defined by

$$f(x; \mu, \sigma^2) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x - \mu}{\sigma}\right)^2}$$

A normal random variable (RV) is not directly integrable, preventing the use of a closed form for its cumulative distribution function (CDF). However, any normal distribution can be effortlessly transformed into a standard normal distribution (where mean = 0 and variance = 1) using the following formula:

$$Z = \frac{{X - \mu}}{{\sigma}}$$


$$P[X \leq x] = P\left[\frac{{X - \mu}}{{\sigma}} \leq \frac{{x - \mu}}{{\sigma}}\right] = P\left[Z \leq \frac{{x - \mu}}{{\sigma}}\right] = \Phi\left(\frac{{x - \mu}}{{\sigma}}\right) = \Phi(z)$$

This `phi` function is a look-up table and also available in python.

**Question:**


Imagine eagerly awaiting a pizza delivery! On average, it takes 22 minutes for the delivery to arrive with a standard deviation of 2 minutes. However, your hunger is compelling you to wonder whether it is possible for the delivery to take more than 30 minutes?

**Answer:**

Let X be the time until delivery

$$X \sim N(\mu=22, \sigma=4)$$

$$P(X>30) = 1 - P(X<30)$$

$$\Phi(\frac{{x - \mu}}{{\sigma}}) = \Phi(\frac{{30 - 22}}{{4}})$$

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
mean, var, x = 22, 4, 30
phiInput = (x - mean)/var

prob = 1 - stats.norm(0, 1).cdf(phiInput)
print("probability for the delivery to take more than 30 minutes: ", prob)
```

```python
xVal = np.linspace(-5,5,100)
yVal = stats.norm(0, 1).pdf(xVal)
```

```python
import matplotlib.pyplot as plt
import numpy as np

xVal = np.linspace(-3, 3, 100)
yVal = np.exp(-0.5 * xVal**2) / np.sqrt(2 * np.pi)

plt.figure()

# Plot the curve
plt.plot(xVal, yVal, color="purple", label='PDF')

# Add a vertical line
plt.axvline(phiInput, color="red", linestyle='--', label='f (X>30)')

# Fill the area under the curve to the right of the line
plt.fill_between(xVal[xVal >= phiInput], yVal[xVal >= phiInput], color='gray', alpha=0.5, label='P (X>30)')

plt.gca().spines["right"].set_visible(False)
plt.gca().spines["top"].set_visible(False)

plt.title("Nromal PDF")
plt.ylabel("f (Z=z)")
plt.xlabel("Z")

plt.ylim(0, 0.45)
plt.legend()
plt.show()
```
