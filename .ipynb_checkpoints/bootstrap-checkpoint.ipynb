# bootstrap

We are usually given a statistic (a single scalar) such as mean and variance of a characteristic in a sample or an accuracy of a machine learning model. However these single number does not represent our uncertainty of the statistic. When dealing with mean, we can use central limit theory to calculate standard error:



$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$

$
S^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})^2
$

$
SE = \frac{S}{\sqrt{n}}
$


But how about variance and other metrics? (central limit theorem can only be applied for average)

Here we can use bootstraping to have a representation of our uncertainty!

**Alg**:
1. Calculate the statistic
2. Resample with replacement for n times
3. Recalculate the statistic and store it

```python
import numpy as np
from scipy import stats 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr = [80, 82, 93, 83, 100, 93, 97, 91, 94, 89, 82, 76, 79, 95, 85, 94, 86]
disMean, disVar, disMode = [], [], []

for i in range(10000):
    
    newSample = np.random.choice(arr, len(arr))
    disMean.append(np.mean(newSample))
    disVar.append(np.var(newSample))
    disMode.append(stats.mode(newSample)[0])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(13,5))

sns.kdeplot(disMean, shade=True, ax=ax[0])
sns.kdeplot(disVar, shade=True, ax=ax[1])
sns.kdeplot(disMode, shade=True, ax=ax[2])


plt.show()
```
